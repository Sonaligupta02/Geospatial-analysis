```python
import pandas as pd
usa_data = pd.read_csv("C:\\Users\\dell\\Downloads\\Task\Florida_Subsidence_Incident_Reports.csv")
usa_data.headad()
```

```python
usa_data = usa_data.iloc[0:100, :]
florida = folium.Map(location= [27.665, -81.516], zoom_start = 7)
incidents = folium.map.FeatureGroup()
for lat, lng, in zip(usa_data.Y, usa_data.X):
    incidents.add_child(
    folium.CircleMarker(
    [lat, lng],
    radius=5,
    color='yellow',
    fill=True,
    fill_color='blue',
    fill_opacity=0.6))
florida.add_child(incidents)
```

```python
latitudes = list( usa_data.Y)
longitudes = list( usa_data.X)
for lat, lng in zip(latitudes, longitudes):
    folium.Marker([lat, lng]).add_to(florida)
florida.add_child(incidents)
```
